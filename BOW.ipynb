```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('quora.csv')
```

```python
# df=pd.read_csv('quora.csv', nrows=30000)-->read only first 30000 rows
```

```python
df.shape
```

```python
df.head()
```

```python
new_df = df.sample(10000)
```

```python
new_df.isnull().sum()
```

```python
new_df.duplicated().sum()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

questions=list(new_df['question1'])+list(new_df['question2'])
# len(questions)-->60000 questions
# new_df['questions']=questions

# new_df.head()
```

```python
cv = CountVectorizer(max_features=10000)
q1_arr, q2_arr = np.vsplit(cv.fit_transform(questions).toarray(),2)
```

```python
temp_df1 = pd.DataFrame(q1_arr, index= new_df.index)
temp_df2 = pd.DataFrame(q2_arr, index= new_df.index)
temp_df = pd.concat([temp_df1, temp_df2], axis=1)
temp_df.shape
```

```python
temp_df
```

```python
temp_df['is_duplicate'] = new_df['is_duplicate']
```

```python
temp_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(temp_df.iloc[:,0:-1].values,temp_df.iloc[:,-1].values,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier()
rf.fit(X_train,Y_train)
Y_pred = rf.predict(X_test)
accuracy_score(Y_test,Y_pred)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train,Y_train)
Y_pred = xgb.predict(X_test)
accuracy_score(Y_test,Y_pred)
```
